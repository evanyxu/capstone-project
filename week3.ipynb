```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from IPython.display import display_html
```

```python
#scrape the Wikipedia page
```

```python
s = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(s,'xml')
table = str(soup.table)
display_html(table,raw=True)
```

```python
#create a dataframe
```

```python
df_toronto = pd.read_html(table)
df_t = df_toronto[0]
headers = ['Postal Code','Borough','Neighborhood']
df_t.columns = headers
df_t.drop(0,axis=0,inplace=True)
df_t
```

```python
#data wrangling
```

```python
df1 = df_t[df_t['Borough'] != 'Not assigned']
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2
```

```python
df2.shape
```

```python
#create a dataframe including geodata
```

```python
location = pd.read_csv('http://cocl.us/Geospatial_data')
location.head()
```

```python
df3 = pd.merge(df2,location,on='Postal Code')
df3.head(10)
```

```python
#cluster the boroughs with 'Toronto'
```

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
# K-means clustering
```

```python
from sklearn.cluster import KMeans
```

```python
k=5
cluster = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(cluster)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```
